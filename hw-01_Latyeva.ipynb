#### **[task 6, 1б.]**
Познакомимся с рынком труда России. Для этого допишите функцию info_count.

```python
def info_count(df, col):
    column = df.get(col)
    g = column.value_counts(sort = True, ascending = True)
    g /= g.sum()
    
    g.plot(kind='barh')
    for index, value in enumerate(g):
        plt.text(value, index, str(round(value*100,1))+'%')
    plt.grid(True)
    plt.show()
```

```python
plot_cols = ['town_type', 
            'professional_group',
            'finished_education', 
            'employed', 'industry',
           ]

for col in plot_cols:
    info_count(df, col)
```

#### **[task 9, 0.5б.]**
Допишите функцию ниже для подсчета [медианной](https://ru.wikipedia.org/wiki/%D0%9C%D0%B5%D0%B4%D0%B8%D0%B0%D0%BD%D0%B0_(%D1%81%D1%82%D0%B0%D1%82%D0%B8%D1%81%D1%82%D0%B8%D0%BA%D0%B0)) [зарплаты](https://www.kommersant.ru/doc/4441454#:~:text=%D0%9C%D0%B5%D0%B4%D0%B8%D0%B0%D0%BD%D0%BD%D0%B0%D1%8F%20%D0%B7%D0%B0%D1%80%D0%BF%D0%BB%D0%B0%D1%82%D0%B0%20%E2%80%94%20%D1%8D%D1%82%D0%BE%20%D1%82%D0%B0%D0%BA%D0%BE%D0%B9%20%D1%83%D1%80%D0%BE%D0%B2%D0%B5%D0%BD%D1%8C,%D1%81%D0%BD%D0%B8%D0%B6%D0%B5%D0%BD%D0%B8%D0%B5%20%D1%81%D1%80%D0%B5%D0%B4%D0%BD%D0%B8%D1%85%20%D0%B7%D0%B0%D1%80%D0%BF%D0%BB%D0%B0%D1%82%20%D0%B2%20%D1%81%D1%82%D1%80%D0%B0%D0%BD%D0%B5.) и построения графика.

```python
def info_wage_median(df, col):    
    groupedColumn = df.groupby(by = col)['wage']
    g = groupedColumn.median().sort_values(ascending = True)
    g = g / 1000
    
    fig, ax = plt.subplots(figsize = (12,6))
    g.plot(kind='barh')
    for index, value in enumerate(g):
        plt.text(value, index, str(int(value)))
    plt.grid(True)
    plt.xlabel('тыс. руб.')
    plt.show()

for col in plot_cols:
    info_wage_median(df_emp, col)
```

### Распределение зарплат

#### **[task 11, 1.5б.]**

Добавить в переменную `df_emp` новый столбец `wage_bin`, где в соответствии с зарплатой будет указан диапазон зарплаты в тыс. рублей. Возможные значения диапазон начинаются от 0 и идут с шагом 4. Например, для зарплаты 2000 рублей должен быть указан диапазон (0,4], а для 5000 рублей – (4,8], и т.д.

```python
maxWage = df_emp['wage'].max() # максимальная зарплата, от неё зависит последний интервал

# округляем максимальную зарплату до ближайшего кратного 4-ём числа
end = (4 - math.floor(maxWage / 1000)) % 4 + math.floor(maxWage / 1000) 

intervals = pd.interval_range(start = 0, end = end, freq = 4)

df_emp['wage_bin'] = pd.cut(df_emp['wage'] / 1000, intervals)
df_emp.head()
```

#### **[task 12, 1б.]**
Посчитайте долю каждого диапазона зарплат в выборке, запишите результат в виде Pandas.Series в переменную g.

```python
g = df_emp["wage_bin"].value_counts(sort = True, ascending = False)
g = g / g.count()

g.head(10)
```

```python
g.plot(kind='bar')
for index, value in enumerate(g):
    plt.text(index, value, str(round(value,2)))
plt.ylabel('frequency')
plt.show()
```

### Gender wage gap

#### **[task 14, 0.5б.]**
Создайте две подвыборки: одну с мужчинами, чья зарплата меньше или равна 100 тыс. руб, и одну с женщинами, чья зарплата также ниже 100 тыс. руб. 

```python
m_under100k = df[(df['wage'] / 1000 <= 100) & (df['sex'] == "М")]
f_under100k = df[(df['wage'] / 1000 <= 100) & (df['sex'] == "Ж")]
```

#### **[task 15, 0.5б]**

Рассмотрите график. Можно ли говорить о значительной разнице в зарплатах между женщинами и мужчинами? 

[Подробнее о феномене](https://en.wikipedia.org/wiki/Gender_pay_gap).

```python
bins=25
maxx=100000

f_under100k['wage'].plot(kind='hist', bins=bins, alpha=0.5, label='Ж')
m_under100k['wage'].plot(kind='hist', bins=bins, alpha=0.5, label='М')
plt.xlim(0,maxx)

plt.legend()
plt.show()
```

Можно увидеть, что большя часть женщин получает сравнительно низкую зарплату, в то же время мужчины уже с отметки в 30000 рублей имееют большую частоту вплоть до самых высоких зарплат
